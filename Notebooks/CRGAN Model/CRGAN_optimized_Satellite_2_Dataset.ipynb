```python

import os
os.environ["CUDA_VISIBLE_DEVICES"] = ""

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Data Preparation for Satellite Dataset
class SatelliteDataset(Dataset):
    def __init__(self, file_path):
        # Load the dataset (adjust encoding as necessary)
        self.data = pd.read_csv(file_path, encoding="ISO-8859-1")

        # Define the columns you want to use
        numeric_columns = ["Longitude of GEO (degrees)", "Eccentricity",
                           "Inclination (degrees)", "Expected Lifetime (yrs.)", "NORAD Number"]
        categorical_columns = ["Country/Org of UN Registry", "Purpose"]
        target_column = "Class of Orbit"

        # Drop rows with missing values in these columns
        self.data = self.data[numeric_columns + categorical_columns + [target_column]].dropna()

        # Process numeric features: scale using StandardScaler
        self.numeric_data = self.data[numeric_columns]
        self.scaler = StandardScaler()
        self.numeric_data = self.scaler.fit_transform(self.numeric_data)

        # Process categorical features: one-hot encode them and record group sizes
        self.categorical_data = pd.get_dummies(self.data[categorical_columns])
        # Calculate group sizes for each categorical column:
        self.cat_group_sizes = [len(pd.get_dummies(self.data[[col]]).columns) for col in categorical_columns]

        # Combine numeric and categorical features
        self.features = np.hstack([self.numeric_data, self.categorical_data.values])

        # Process target: encode target labels as integers
        self.classes = sorted(self.data[target_column].unique())
        class_to_idx = {c: i for i, c in enumerate(self.classes)}
        self.labels = self.data[target_column].map(class_to_idx).values

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return torch.FloatTensor(self.features[idx]), torch.LongTensor([self.labels[idx]])

# Set the Satellite file path
file_path_sat = "/content/drive/MyDrive/Katabatic/Data/Satellite/Satellite-Database.csv"
dataset_sat = SatelliteDataset(file_path_sat)

# Split the Dataset
# Split the dataset into train (80%) and test (20%) subsets using random_split
train_size = int(0.8 * len(dataset_sat))
test_size = len(dataset_sat) - train_size
train_dataset, test_dataset = random_split(dataset_sat, [train_size, test_size])

# Create DataLoader for training
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True)

#Define Modified Generator & Discriminator
# Modified SatelliteGenerator with separate activations for numeric and categorical parts
class SatelliteGenerator(nn.Module):
    def __init__(self, latent_dim, numeric_dim, cat_group_sizes):
        super(SatelliteGenerator, self).__init__()
        self.numeric_dim = numeric_dim
        self.cat_group_sizes = cat_group_sizes
        self.total_cat_dim = sum(cat_group_sizes)
        self.total_output_dim = numeric_dim + self.total_cat_dim

        self.fc = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, self.total_output_dim)  # Raw output
        )

    def forward(self, z):
        out = self.fc(z)  # Shape: (batch_size, total_output_dim)
        # Numeric part: first self.numeric_dim elements, apply tanh
        numeric_out = torch.tanh(out[:, :self.numeric_dim])
        # Categorical part: remaining elements, split into groups and apply softmax to each group
        cat_raw = out[:, self.numeric_dim:]
        cat_outputs = []
        start = 0
        for size in self.cat_group_sizes:
            group = cat_raw[:, start:start+size]
            group_softmax = torch.softmax(group, dim=1)
            cat_outputs.append(group_softmax)
            start += size
        cat_out = torch.cat(cat_outputs, dim=1)
        # Concatenate numeric and categorical outputs
        final_out = torch.cat([numeric_out, cat_out], dim=1)
        return final_out

# Discriminator remains as before
class Discriminator(nn.Module):
    def __init__(self, input_dim):
        super(Discriminator, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.fc(x)

# Get dimensions
numeric_dim = 5  # As defined in numeric_columns
cat_group_sizes = dataset_sat.cat_group_sizes  # e.g., for ["Country/Org of UN Registry", "Purpose"]
total_feature_dim = numeric_dim + sum(cat_group_sizes)

# --------------------- 4. Initialize Models ---------------------
latent_dim = 100
generator_sat = SatelliteGenerator(latent_dim, numeric_dim, cat_group_sizes)
discriminator_sat = Discriminator(total_feature_dim)

# --------------------- 5. Train CRAMERGAN ---------------------
def train_cramer_gan(generator, discriminator, dataloader, epochs=100):
    device = torch.device('cpu')  # Force CPU usage for stability
    generator.to(device)
    discriminator.to(device)

    optimizer_g = optim.Adam(generator.parameters(), lr=0.0002)
    optimizer_d = optim.Adam(discriminator.parameters(), lr=0.0002)
    scheduler_g = optim.lr_scheduler.StepLR(optimizer_g, step_size=20, gamma=0.5)
    scheduler_d = optim.lr_scheduler.StepLR(optimizer_d, step_size=20, gamma=0.5)
    criterion = nn.BCELoss()

    for epoch in range(epochs):
        for real_data, _ in dataloader:
            real_data = real_data.to(device)
            batch_size = real_data.size(0)

            # Train Discriminator
            optimizer_d.zero_grad()
            z = torch.randn(batch_size, latent_dim).to(device)
            fake_data = generator(z).detach()
            real_labels = torch.ones(batch_size, 1).to(device)
            fake_labels = torch.zeros(batch_size, 1).to(device)
            loss_real = criterion(discriminator(real_data), real_labels)
            loss_fake = criterion(discriminator(fake_data), fake_labels)
            loss_d = loss_real + loss_fake
            loss_d.backward()
            optimizer_d.step()

            # Train Generator
            optimizer_g.zero_grad()
            z = torch.randn(batch_size, latent_dim).to(device)
            fake_data = generator(z)
            loss_g = criterion(discriminator(fake_data), real_labels)
            loss_g.backward()
            optimizer_g.step()

        scheduler_g.step()
        scheduler_d.step()
        print(f"Epoch [{epoch+1}/100] - Loss D: {loss_d.item():.4f}, Loss G: {loss_g.item():.4f}")

# Train CRAMERGAN on the training set of the Satellite dataset
train_cramer_gan(generator_sat, discriminator_sat, train_dataloader, epochs=100)

# Generate Synthetic Data & Post-Process Categorical Features
def generate_synthetic_data(generator, num_samples, latent_dim):
    device = torch.device('cpu')
    generator.eval()
    z = torch.randn(num_samples, latent_dim).to(device)
    synthetic_data = generator(z).detach().cpu().numpy()
    return synthetic_data

# Generate synthetic samples equal to the number of training samples
num_synthetic_samples = len(train_dataset)
synthetic_raw = generate_synthetic_data(generator_sat, num_synthetic_samples, latent_dim)

# Post-processing:
# Split synthetic output into numeric and categorical parts
synthetic_numeric = synthetic_raw[:, :numeric_dim]  # Numeric part
synthetic_cat_raw = synthetic_raw[:, numeric_dim:]    # Categorical probabilities

# For each categorical group, apply argmax then convert to one-hot encoding
def post_process_categorical(synthetic_cat, group_sizes):
    processed_groups = []
    start = 0
    for size in group_sizes:
        group = synthetic_cat[:, start:start+size]  # shape: (num_samples, group_size)
        # Argmax to get discrete class labels
        discrete_labels = np.argmax(group, axis=1)  # shape: (num_samples,)
        # Convert discrete labels back to one-hot vectors
        one_hot = np.eye(size)[discrete_labels]  # shape: (num_samples, group_size)
        processed_groups.append(one_hot)
        start += size
    return np.hstack(processed_groups)

synthetic_cat_processed = post_process_categorical(synthetic_cat_raw, cat_group_sizes)

# Inverse transform the numeric part back to original scale
synthetic_numeric = dataset_sat.scaler.inverse_transform(synthetic_numeric)

# Combine numeric and processed categorical parts
synthetic_data = np.hstack([synthetic_numeric, synthetic_cat_processed])

# Print the number of synthetic samples generated
print("Number of synthetic samples generated:", synthetic_data.shape[0])

# --------------------- 7. Evaluate Models on Real Data ---------------------
# For evaluation, load the real Satellite data (the numeric+one-hot features from our dataset)
# Extract training data from train_dataset indices
train_indices = train_dataset.indices
X_train_real = dataset_sat.features[train_indices]
y_train_real = np.array(dataset_sat.labels)[train_indices]

# Extract test data from test_dataset
test_indices = test_dataset.indices
X_test = dataset_sat.features[test_indices]
y_test = np.array(dataset_sat.labels)[test_indices]

# We train on synthetic data only, then evaluate on real test data.
# Alternatively, you could mix synthetic and real training data.
models = {
    "Logistic Regression": LogisticRegression(max_iter=5000, solver='lbfgs', C=1.0),
    "MLP": MLPClassifier(hidden_layer_sizes=(128,64), max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=200),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', n_estimators=200, max_depth=6, learning_rate=0.1)
}

results = []
print("\nClassifier Evaluation Results (trained on synthetic data, tested on real test set):")
for name, model in models.items():
    model.fit(synthetic_data, y_train_real)  # Train on synthetic data with real training labels
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results.append((name, acc))
    print(f"{name}: {acc:.4f}")

results_df = pd.DataFrame(results, columns=["Model", "Accuracy"])
print("\n", results_df)

# --------------------- 8. Visualize & Save Synthetic Data ---------------------
# Create a DataFrame for synthetic data with proper column names.
# Numeric columns labeled as "Numeric i" and categorical columns as "Cat i"
numeric_columns = [f"Numeric {i}" for i in range(numeric_dim)]
categorical_columns = []
start = 0
for i, size in enumerate(cat_group_sizes):
    # Create column names for each categorical group, e.g., "Cat 0", "Cat 1", ..., one per one-hot dimension
    categorical_columns.extend([f"Cat {i}_{j}" for j in range(size)])
all_columns = numeric_columns + categorical_columns

synthetic_df = pd.DataFrame(synthetic_data, columns=all_columns)
print("\nSynthetic Data Sample:")
print(synthetic_df.head())

# Save synthetic data to a CSV file in your Satellite folder on Google Drive
csv_save_path = "/content/drive/MyDrive/Katabatic/Data/Satellite/synthetic_satellite_data.csv"
synthetic_df.to_csv(csv_save_path, index=False)
print("\nSynthetic data saved to:", csv_save_path)
```
